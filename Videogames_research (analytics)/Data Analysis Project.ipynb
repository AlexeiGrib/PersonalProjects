### Этап 2.2.3. Обработка аномальных значений в поле User_Score.

Напомним, что в этом поле мы выявлили аномальную запись tbd.

Для начала посмотрим, что общего у игр, для которых в поле User_Score указано значение tbd.

```python
df.query('User_Score == "tbd"')
```

Видим, что таких записей у нас 859, что составляет чуть больше 5% от общего объёма df. Цифра значительная, поэтому удалять такие поля не станем.

Вместо этого проанализируем уникальные значения в таких полях, как Platform, Year_of_Release и Genre.

```python
for col in df.query('User_Score == "tbd"').columns:
    if col in ('Platform', 'Year_of_Release', 'Genre'):
        print(col)
        print()
        print(df[col].value_counts())
        print()
```

Как видим, между играми, у которых в поле User_Score указано значение tbd, нет ничего общего: для них характерны разные годы выпуска, разные платформы и разные жарны.

Вместе с тем в игровой индустрии есть аббревиатура tbd (To Be Determined), что означает, что какой-то параметр, которому присвоена данная аббревиатура, будет определён позже. Обычно это характерно для дат релиза игр или дат получения оценок критиков или игроков.

Таким образом, приходим к выводу о том, что запись в поле User_Score со значением tbd является аналогом значения Nan. 
Для преобразования типа данных из object в float и обработки аномальных значений tbd воспользуемся методом Pandas to_numeric. При этом исключения будем преобразовывать в NaN, для этого необходимо в передать параметру errors аргумент coerce: тогда все исключения будут заменены на NaN.

```python
df['User_Score'] = pd.to_numeric(df['User_Score'], errors = 'coerce')
```

### Этап 2.3.4. Обработка пропущенных значений в поле Critic_Score и User_Score.

Пропущенные значения в этом поле можно заменить на медианное значение оценки критиков в зависимости от жанра.

Для этого создадим сводную таблицу, сгруппируем данные по жанрам, посчитаем медианную оценку критиков для каждого из жанров и сформируем из этой таблицы словарь.

### Этап 2.3.6. Обработка пропущенных значений в поле Rating.

Пропущенные значения в этом поле, вероятно, имеют ту же природу, что и в поле Crtitc_Score: данные просто не были получены (возможно, проблема с выгрузками или загрузкой информации из внешних источников).

Однако логика обработки пропусков, применяемая в предыдущих этапах, здесь не может быть применена: оценки здесь имеют строковый формат, а не числовой, следовательно, медиану мы не посчитаем. Поэтому немного поизвращаемся.

Сначала создадим словарь, в котором категоризированные значения поля Rating будут ключом, а значением - его цифровой аналог.

Отмечу также, что мы не будем исключать пропущенные значения, так как потом не сможем сделать замену - будут мешать пропущенные значения (или nan).

```python
dict_median_user_rating = dict()
i = 0
for value in df['Rating']:
    if value not in dict_median_user_rating:
        i += 1
        dict_median_user_rating[value] = i
```

Посмотрим на получившийся словарь.

```python
dict_median_user_rating
```

Видим, что мы успешно создали словарь из значений поля Rating и присвоили им соответствующее цифровое обозначение.

Далее напишем функцию, которая будет принимать на вход значения из поля Rating и возвращать соответствующее значение из словаря, используя значения поля Rating как ключ.

```python
def rating_value(Rating):
    return dict_median_user_rating[Rating]
```

Функция написана. Проверим её работу на тестовых данных.

```python
print(rating_value('E'))
print(rating_value('M'))
print(rating_value('EC'))
print(rating_value('RP'))
print(rating_value('AO'))
```

Видим, что функция хорошо работает на тестовых данных.

Применим функцию к нашему набору данных. Результаты работы функции поместим в новое поле Rating_Value.

```python
df['Rating_Value'] = df['Rating'].apply(rating_value)
```

Проверим работу функции.

```python
df.head(10)
```

Видим, что функция успешно поместила в поле Rating_Value цифровое значение рейтинга.

Далее посчитаем медианное цифровое значение рейтинга для каждого жанра. При этом значение рейтинга 2 (что соответствует пропущенному значению) учитывать не будем по очевидным причинам.

Результаты передадим в словарь.

```python
dict_median_user_rating_values = df.query('Rating_Value != 2').pivot_table(index = 'Genre', values = 'Rating_Value', 
aggfunc = 'median')['Rating_Value'].to_dict()
dict_median_user_rating_values
```

Видим, что мы успешно посчитали медианную цифровую оценку рейтинга игры для каждого жанра. Теперь заменим в поле Rating_Value значения 2 (которые соответствуют пропущенным значениям в поле Rating) на соответствующее жанру игры медианное значение цифровой оценки рейтинга. Для этого напишем функцию, которая будет принимать на вход строку, сравнивать значение поля Rating_Value с значением 2 и реализовывать соответствующую логику замены.

```python
def rating_value_replace(row):
    Rating_Value = row['Rating_Value']
    Genre = row['Genre']
    if Rating_Value == 2:
        return dict_median_user_rating_values[Genre]
    else:
        return Rating_Value
```

Применим функцию. Так как нам не нужно создавать новое поле, результаты работы функции будем присваивать значениям уже существующего поля Rating_Value.

```python
df['Rating_Value'] = df.apply(rating_value_replace, axis = 1)
```

Посмотрим, корректно ли осуществлена замена.

```python
df.head(10)
```

Видим, что замена осуществлена успешно. Теперь, чтобы заменить пропущенные значения в поле Rating, напишем функцию, которая будет принимать на вход словарь и значение, а возвращать будет ключ для этого значения.

```python
def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k
```

Проверим работу функции на тестовых данных.

```python
print(get_key(dict_median_user_rating, 1))
print(get_key(dict_median_user_rating, 2))
print(get_key(dict_median_user_rating, 3))
print(get_key(dict_median_user_rating, 6))
print(get_key(dict_median_user_rating, 8))
```

Видим, что функция работает корректно на тестовых данных. Теперь, чтобы завершить работу над пропущенными значениями в поле Rating, напишем функцию, которая будет принимать на вход строку, искать в нём значение в поле Rating_Value и возвращать в поле Rating ключ (соответствующий рейтингу) из словаря dict_median_user_rating, а в качестве значения принимать значение из поля Rating_Value.

```python
def from_value_in_key(row):
    Rating_Value = row['Rating_Value']
    return get_key(dict_median_user_rating, Rating_Value)
```

Применим функцию к нашему набору данных.

```python
df['Rating'] = df.apply(from_value_in_key, axis = 1)
```

Проверим, корректно ли осуществлена замена.

```python
df.head(10)
```

Видим, что пропущенных значений в df не осталось - мы успешно их обработали.

Результаты визуализируем с помощью гистограммы: по оси x отложим год релиза игры, по оси y - суммарный объём продаж. 

Отмечу, что перед использованием сводной таблицы df будет ограничен по полю Platform, чтобы исследовать 1 платформу за раз, а также мы будем рассматривать игры с датой релиза равной или старше 2000, так как на прошлом этапе мы определили, что для анализа целесообразно использовать данные с 2000 года и до последних имеющихся в наличии сведений.

Для автоматизации построения графиков воспользуемся циклом. 

Введём список platforms, в котором будут храниться уникальные наименования платформ, для которых выходили игры с датой релиза равной или старше 2000 года. Соответственно, при работе цикла будем перебирать платформы из этого списка.

```python
platforms = list(df.query('Year_of_Release >= 2000')['Platform'].unique())
for platform in platforms:
    df[(df['Platform'] == platform) & (df['Year_of_Release'] >= 2000)].pivot_table(index = ['Platform', 'Year_of_Release'], 
    values = 'Total_sales', aggfunc = 'sum').reset_index().plot(y = 'Total_sales', x = 'Year_of_Release', 
    kind = 'bar', grid = True,
    legend = False, figsize = (20, 5))
    plt.title('Суммарный объём продаж игр для платформы ' + str(platform))
    plt.xlabel('Год релиза игр')
    plt.ylabel('Суммарный объём продаж, млн. долларов')
    plt.show()
```

```python
df[df['Platform'] == "PC"].pivot_table(index = ['Platform', 'Year_of_Release'], 
values = 'Total_sales', aggfunc = 'sum').reset_index().plot(y = 'Total_sales', x = 'Year_of_Release', 
kind = 'bar', grid = True,
legend = False, figsize = (20, 5))
plt.title('Суммарный объём продаж игр для платформы PC')
plt.xlabel('Год релиза игр')
plt.ylabel('Суммарный объём продаж, млн. долларов')
plt.show()
```

## Этап 3.5. Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков? Как можно соотнести полученные выводы с продажами игр на других платформах?

Чтобы ответить на этот вопрос, сначала определимся, какую платформу будем исследовать.

На этапе 3.3.4. мы исследовали суммарные продажи по платформам с 2008 года, и наибольший объём продаж был замечен за платформой PS3 (Ваш покорный слуга, как фанат PC, тяжко вздохнул).

Поэтому исследовать будем именно PS3. Для анализа зависимости будем использовать диаграмму рассеяния и коэффициенты корреляции.

```python
df[df['Platform'] == 'PS3'].plot(x = 'User_Score', y = 'Total_sales', kind = 'scatter')
df[df['Platform'] == 'PS3'].plot(x = 'Critic_Score', y = 'Total_sales', kind = 'scatter')
plt.show()
```

Пока что диаграмма рассеяния выглядит не очень презентабельно, и виной тому - выбросы в поле Total_sales: выбросы сосредоточены в праовой части диаграммы и довольно малочисленны, а остальные значения сбились в огромное неразличимое пятно в левой части диаграммы. 

Построим эти диаграммы заново, но уже без выбросов.

```python
df[(df['Platform'] == 'PS3') & (df['Total_sales'] <= df[df['Platform'] == 'PS3']['Total_sales'].quantile(0.95))].\
plot(x = 'Critic_Score', y = 'Total_sales', kind = 'scatter')
df[(df['Platform'] == 'PS3') & (df['Total_sales'] <= df[df['Platform'] == 'PS3']['Total_sales'].quantile(0.95))].\
plot(x = 'User_Score', y = 'Total_sales', kind = 'scatter')
plt.show()
```

Теперь лучше. Видим, что оценки - как критиков, так и игроков - не оказывают значительного влияния на суммарные продажи: например, значительная часть продаж лежит в диапазоне от 0.0 млн. долларов до 0.5 млн. долларов, однако для игр, формировавших эти продажи, оценки лежат в разном диапазоне - от очень высоких до очень низких.

Поэтому рабочая гипотеза - оценки не влияют на объёмы продаж.

Чтобы закрепить гипотезу, посчитаем коэффициенты корреляции.

```python
df[(df['Platform'] == 'PS3') & (df['Total_sales'] <= df[df['Platform'] == 'PS3']['Total_sales'].quantile(0.95))]\
[['Critic_Score', 'User_Score', 'Total_sales']].corr().tail(1)
```

Так как значения коэффициента корреляции между изучаемыми признаками всегда больше нуля, можно говорить о том, что линейная зависимость - какой бы сильной она ни была - имеет прямую направленность, т.е. по мере росто какого-либо из признаков будет расти и суммарный объём продаж. 

В своей практике я применяю следующую систему критериев оценки силы линейной зависимости:

- линейная зависимость сильная, если значение коэффициента корреляции лежит в пределах от ±0.70 до ±1.00;

- линейная зависимость средняя, если значение коэффициента корреляции лежит в пределах от ±0.30 до ±0.69;

- линейная зависимость слабая, если значение коэффициента корреляции лежит в пределах от 0.00 до ±0.29.

Полученные нами значения коэффициента корреляции предлагаю интерпретировать таким же образом.

Так, зависимость между:

- суммарным объёмом продаж и оценкой критиков - средняя прямая; 

- суммарным объёмом продаж и оценкой игроков - низкая прямая.

Отмечу также, что коэффициент корреляции не позволяет доподлинно утверждать о наличии или отсутствии зависимости между изучаемыми показателями, а лишь показывает наличие или отсутствие линейной связи между показателями, а также тесноту этой связи, в рамках изучаемой выборки.

Так, коэффициент корреляции указал на среднюю зависимость между суммарными продажами и оценками критиков и низкую зависимость между суммарными продажами и оценками пользователей. 

Перенося этот вывод на практику, в целом, это не лишено здравого смысла - в основном игроки не обращают внимание на оценки критиков/игроков при принятии решения о покупке игры. Более важными факторами может быть имя компании-разработчика, любимый жанр, атмосфера игры, преданность покупателя определённой серии игр и др.

Посмотрим, справедливы ли данные выводы для других платформ.

```python
print()
for i in list(df['Platform'].unique()):
    print('Зависимость суммарных продаж от оценки критиков/игроков для платформы ' + str(i))
    display(df[(df['Platform'] == i) & (df['Total_sales'] <= df[df['Platform'] == i]['Total_sales'].quantile(0.95))]\
    [['Critic_Score', 'User_Score', 'Total_sales']].corr().tail(1))
    print()
```

В целом можно отметить, что выводы о зависимости суммарного объёма продаж игр и оценок критиков/игроков для платформы PS3 справедливы и для остальных платформ:

Видим, что Северная Америка лидирует по суммарному объёму продаж, поэтому этот регион и выберем в качестве объекта исследования.

Чтобы оценить влияние рейтинга на продажи, можно воспользоваться коэффициентом корреляции и диаграммой рассеяния.

Отмечу, что ранее рейтинг был приведён в буквенном формате, однако мы предусмотрительно создали новое поле с его численным аналогом, так что теперь можем без труда посчитать корреляцию и построить диаграмму рассеяния.

```python
df.plot(x = 'NA_sales', y = 'Rating_Value', kind = 'scatter')
plt.show()
```

На первый взгляд линейной зависимости между переменными нет. Чтобы удостовериться в этом, рассчитаем коэффициент корреляции.

```python
print('Коэффициент корреляции объёмов продаж и рейтинга:', df['NA_sales'].corr(df['Rating_Value']))
```

Коэффициент корреляции также говорит о том, что между переменными нет значимой линейной зависимости. Однако коэффициент корреляции не может описать более сложную зависимость, которая имеет нелинейный характер.

Эмпирически очевидно, что рейтинг так или иначе может влиять на продажи игры. Поэтому сгруппируем продажи по рейтингу и посмотрим, что получится.

```python
df.pivot_table(index = 'Rating', values = 'NA_sales', aggfunc = 'sum').reset_index().plot(x = 'Rating', y = 'NA_sales', 
kind = 'bar', grid = True, legend = False, figsize = (20, 5))
plt.xlabel('Рейтинг')
plt.ylabel('Объём продаж (в млн. долларов)')
plt.show()
```

Из данного графика видим, что игры с рейтингом EC, К-А и RP не пользуются большой популярностью, в то время как игры с рейтингом E и М бьют рекорды продаж и завоёвывают сердца игроков. Рейтинги E10+ и T заняли золотую серидину и чувствуют себя вполне хорошо - наелись и спят.

# Этап 5. Построение гипотез и их проверка.

На данном этапе нужно проверить или опровергнуть следующие **гипотезы:**

## Этап 5.1. Построение гипотезы "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые" и её проверка.

Все необходимые данные содержатся в df.

Для проверки гипотезы нам нужны 2 среза: с оценками игроков для платформы XBox One и PC. Также для проверки гипотезы нам понадобится преобразование Series в массив данных Array. Используем метод values.

```python
users_score_xone = df.query('Platform == "XOne" and Year_of_Release >= 2008 and User_Score.notna()')['User_Score'].values
users_score_pc = df.query('Platform == "PC" and Year_of_Release >= 2008 and User_Score.notna()')['User_Score'].values
```

Видим, что мы успешно извлекли из df оценки игроков для каждой платформы и преобразовали эти данные в массив. 

Формулируем нулевую гипотезу H0: средние пользовательские рейтинги платформ Xbox One и PC одинаковы.
    
Формулируем альтернативную гипотезу H1: средние пользовательские рейтинги платформ Xbox One и PC различаются.

Нулевая гипотеза формируется таким образом, чтобы результат проверки был для нас ожидаем и неизменен: если мы проверям отличие между двумя средними, нулевая гипотеза должна быть сформирована таким образом, что различий между средними не должно быть.

Условимся, что будем использовать уровень значимости 0.05 для проверки гипотез.

Теперь необходимо выбрать подходящий критерий проверки гипотезы. Для этого нужно понять, с каким распределением мы имеем дело - нормальным или отличным от нормального.

Чтобы это проверить, воспользуемся normaltest из библиотеки scipy.stats.

Гипотеза H0 - данные в выборке распределены нормально.

Гипотеза H1 - данные в выборке распределены как-то иначе.

Проверим, нормально ли распределены данные в users_score_xone.

```python
alpha = 0.05
results = st.normaltest(users_score_xone)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза отклонена: распределение данных не является нормальным.')
else:
    print('Нулевая гипотеза подтверждена: данные распределены нормально.')
```

То же проделаем и для users_score_pc.

```python
alpha = 0.05
results = st.normaltest(users_score_pc)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза отклонена: распределение данных не является нормальным.')
else:
    print('Нулевая гипотеза подтверждена: данные распределены нормально.')
```

Проверка гипотез показала, что при заданном уровне значимости распределение данных отличается от нормальных.

Таким образом, мы имеем 2 независимые друг от друга группы, данные в которых не подчиняются нормальному закону распределения. Следовательно, для проверки гипотезы о равенстве или различии средних пользовательских рейтингов платформ Xbox One и PC используем критерий Манна-Уитни.

```python
alpha = 0.05
results = st.mannwhitneyu(users_score_xone, users_score_pc)
print('Среднее значение пользовательской оценки для платформы XBox One:', round(np.mean(users_score_xone), 2))
print('Среднее значение пользовательской оценки для платформы PC:', round(np.mean(users_score_pc), 2))
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Нулевая гипотеза отклонена: средние пользовательские рейтинги платформ Xbox One и PC различаются.')
else:
    print('Нулевая гипотеза подтверждена: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.')
```

Нулевая гипотеза отклонена: оценки по платформам платформ Xbox One и PC различаются. 

И ведь действительно различаются: учитывая, что верхняя граница оценки - 10, даже маленькая разбежка в оценках может по-разному описывать качество игры.
